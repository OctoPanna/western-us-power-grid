```python
G = load_data_with_attributes()
```

```python
draw_graph(G)
```

```python
# Degree histogram
degree_hist = nx.degree_histogram(G)
degrees = range(len(degree_hist))
counts = degree_hist

plt.figure(figsize=(8, 6))
plt.bar(
    degrees,
    counts, 
    width=0.80, 
    color='#ffc62a', 
    edgecolor='black'
)

plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Number of Nodes")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(left=0)
plt.show()
```

```python
degrees = [G.degree(n) for n in G.nodes()]
capacities = [G.nodes[n]["capacity"] for n in G.nodes()]
loads = [G.nodes[n]["load"] for n in G.nodes()]
load_capacity_ratios = [loads[i] / capacities[i] if capacities[i] > 0 else 0 for i in range(len(capacities))]
```

```python
plt.scatter(degrees, capacities, alpha=0.6)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(left=0)
plt.xlabel("Degree")
plt.ylabel("Capacity")
plt.title("Degree vs Capacity")
plt.show()
```

```python
plt.hist(capacities, bins=30, edgecolor='black')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(left=0)
plt.title("Capacity Distribution")
plt.xlabel("Capacity")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(loads, bins=30, edgecolor='black')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(left=0)
plt.title("Load Distribution")
plt.xlabel("Load")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(load_capacity_ratios, bins=30, edgecolor='black')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(left=0)
plt.title("Load Capacity Ratio Distribution")
plt.xlabel("Load Capacity Ratio")
plt.ylabel("Frequency")
plt.show()
```
